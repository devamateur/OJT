**한 번 치환된 값은 다시 치환되지 않음**
 


```python
dictionary = {"ab":"xxx", "cde":"yyy", "rr":"QQ", "cdefgh":"zzz", "z":"X"}
src = "qqq abcde rrr abcdefghz"
isReplaced = [0]*len(src)          # 치환 여부 확인 리스트

# 1. dictionary key 기준으로 내림차순 정렬
new_dict = dict(sorted(dictionary.items(), key=lambda x: len(x[0]), reverse=True))
print(new_dict)
```

```python
result = src
substring = ""

for key, value in new_dict.items():        # 딕셔너리 key, value
    substring=""

    rid = 0
    for i in range(len(isReplaced)):
        print(i)
        if src[i] != ' ':
            substring += src[i]
            print('substring: ', substring)
            print('key: ', key)
            
        # 딕셔너리 key값과 같고, 교체된 적이 없으면
        if (isReplaced[i] == 0) and (substring == key):
            result = result.replace(substring, value)
            isReplaced[i-len(substring)+1:i+1]=[1]*len(substring)
            
            # isReplaced resize
            if len(key) > len(value):
                isReplaced = isReplaced[:i-len(substring)+len(value)+1] + isReplaced[i+1:]
                print(f'isReplaced 수정(축약): {isReplaced} 길이:{len(isReplaced)}')
                #i=i-len(substring)+len(value)
            elif len(key) < len(value):
                for j in range(len(value)-len(key)):
                    isReplaced.insert(i+j, 1)
                print(f'isReplaced 수정(추가): {isReplaced} 길이:{len(isReplaced)}')
                #i=i+len(value)-len(key)
                
            print('key == substring', substring)
            substring = ""
            
        # substring과 key가 같지 않으면 substring을 뒤로 한 칸씩 민다    
        if (substring != key) and (len(substring) == len(key)):
            substring = substring[1:]
            
        if i >= len(isReplaced):
            break
    else:
        if(len(key)>len(value)):
            i=i-len(substring)+len(value)
        elif len(key) < len(value):
            i=i+len(value)-len(key)
```
