**한 번 치환된 값은 다시 치환되지 않음**
 


```python
dictionary = {"ab":"xxx", "cde":"yyy", "rr":"QQ", "cdefgh":"zzz", "z":"X"}
src = "qqq abcde rrr abcdefghz"
isReplaced = [0]*len(src)          # 치환 여부 확인 리스트

# key 기준으로 내림차순 정렬
new_dict = dict(sorted(dictionary.items(), key=lambda x: len(x[0]), reverse=True))
print(new_dict)
```

```python
result = src
substring = ""

for key, value in new_dict.items():        # 딕셔너리 key, value
    substring=""

    for i in range(len(result)):
        
        print(f'result: {len(result)} {result}')
        print(f'isReplaced: {len(isReplaced)} {isReplaced}\n')
        
        if len(substring) < len(key):
            substring += result[i]
        print(f'>>>> substring: {substring}, 인덱스: {i}, 치환 여부: {isReplaced[i]}')
        print('>>>> key: ', key)
        print()
            
        # 딕셔너리 key값과 같고, 교체된 적이 없으면
        if (isReplaced[i] == 0) and (substring == key):
            print('***** key == substring *****')
            result = result[:i-len(substring)+1] + value + result[i+1:]
            print(f'"{key}"를 "{value}"로 교체 후 result: {result}\n')
            isReplaced[i-len(substring)+1:i+1]=[1]*len(substring)         # 치환 여부 true(1)로 변경
            
            # isReplaced resize - key와 value의 길이가 다를 때, 
            if len(key) > len(value):
                isReplaced = isReplaced[:i-len(substring)+len(value)+1] + isReplaced[i+1:]
                print(f'isReplaced 수정(축약): {isReplaced} 길이:{len(isReplaced)}')
            elif len(key) < len(value):
                print('insert 전 isReplaced ', isReplaced)
                for j in range(1, len(value)-len(key)+1):
                    isReplaced.insert(i+j, 1)
                print(f'insert 후 isReplaced: {isReplaced} 길이:{len(isReplaced)}')
            substring = ""
            
        # substring과 key가 같지 않으면 substring을 뒤로 한 칸씩 민다    
        if (substring != key) and (len(substring) == len(key)):
            substring = substring[1:]
        
        # 치환되면서 result의 길이가 짧아지는 경우가 발생
        if i >= len(result):
            #print('break', i, len(result))
            break
'''    else:
        #if len(key)>len(value):
        #    i=i-len(substring)+len(value)
        if len(key) < len(value):
            i=i+len(value)-len(key)'''
```
